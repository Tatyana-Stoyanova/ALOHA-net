Plotten Sie den Zusammenhang zwischen die Netzbelastung und den relativen Anteil von erfolgreich gesendeten Paketen für einen ALOHA-Netz mit feste Paketgröße mit p = 0,1 und p=0,5.

```python
# Offered load (rate) G = number of (old and new) messages that contend for the channel per time unit
# Throughput S = number of messages that are successfully transmitted per time unit
# S = G*e^(-2G)
```

```python
import numpy as np
import matplotlib.pyplot as pl
import math
```

```python
# S = G*e^(-2G) 
G = np.linspace(0,10,100)
G
S=[]
for i in range(len(G)):
  S.append(G[i]*math.exp(-G[i]*2))
```

```python
maxS=S[0]
maxG=G[0]
for i in range(1,len(S)):
  if(S[i]>=S[i-1]):
    maxS=S[i]
    maxG=G[i]
```

```python
pl.plot(G, S)
pl.plot(maxG,maxS,'bo')
```

```python
# p = Sendewahrscheinlichkeit
# n = Anzahl von Kanälen

p1 = 0.1
p2 = 0.5
n = np.arange(0,100,1)
```

```python
g1 = n*p1
g2 = n*p2
```

```python
prob1=[]
for i in range(len(g1)):
  prob1.append(g1[i]*math.exp(-2*g1[i]))

prob2=[]
for j in range(len(g2)):
  prob2.append(g2[j]*math.exp(-2*g2[j]))
```

```python

pl.plot(g1, prob1)
```

```python
pl.plot(g2, prob2)
```
